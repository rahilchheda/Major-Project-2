### Detecting the outliers in column Kilometers

```python
z_scores = np.abs((df['Kilometers'] - df['Kilometers'].mean()) / df['Kilometers'].std())

# Define threshold for outliers (e.g., Z-score > 3)
threshold = 3
outliers = df[z_scores > threshold]

print("Outliers detected by Z-score method in Kilometers column:")
print(outliers)
```

```python
df_no_outliers = df[z_scores <= threshold]

# Display DataFrame after removing outliers
print("\nDataFrame after removing outliers:")
print(df_no_outliers)
```

## Data Visualization

```python

# Plotting the barplot
plt.figure(figsize=(4,6))
sns.barplot(data=df, x='Cars', y='Price(Lakh)')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to make room for rotated labels
plt.show()
```

```python

# Plotting the barplot
plt.figure(figsize=(4,6))
sns.barplot(data=df, x='Cars', y='Transmission')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to make room for rotated labels
plt.show()
```

# Stacked Bar chart

```python
transmission_counts = df.groupby(['Cars', 'Transmission']).size().reset_index(name='Counts')

# Pivot the data to make 'Transmission' types as columns
transmission_pivot = transmission_counts.pivot(index='Cars', columns='Transmission', values='Counts').fillna(0)

# Plotting stacked bar chart
plt.figure(figsize=(10, 6))
transmission_pivot.plot(kind='bar', stacked=True)
plt.title('Transmission Distribution by Car')
plt.xlabel('Cars')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.legend(title='Transmission')
plt.tight_layout()
plt.show()
```

# Scatter plot

```python
plt.figure(figsize=(4, 4))  # Adjust the figure size to make the graph smaller

plt.scatter(df['Cars'], df['Price(Lakh)'])

plt.xlabel('Cars')
plt.ylabel('Price')

plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.grid(False, axis='x')  # Remove the grid lines on the x-axis

plt.show()
```
